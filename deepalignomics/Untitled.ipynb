```python
from viz import show_alignment
import numpy as np
from sklearn.manifold import TSNE
```

```python
proj_outputs_np = np.load('./data/proj_outputs_np.npy')
```

```python
show_alignment(proj_outputs_np[0:1768], proj_outputs_np[1768:3536])
```

```python
X_embedded = TSNE(n_components=3, perplexity=50).fit_transform(proj_outputs_np[0:1768])
Y_embedded = TSNE(n_components=3, perplexity=50).fit_transform(proj_outputs_np[1768:3536])
show_alignment(X_embedded, Y_embedded, 'Ex1', 'In1')
```
